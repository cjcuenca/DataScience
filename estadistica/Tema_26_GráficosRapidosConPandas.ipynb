Imaginamos otro df

```python
df_2 = pd.DataFrame({"X": [10, 20, 30, 40, 50, 60, 70, 80],
"y": [10, 20, 10, 40, 20, 30, 20, 10]})
df_2
```

```python
df_2.y.value_counts()
```

```python
df_2.y.value_counts().plot(kind="bar")
plt.show()
```

Con color diferente...

```python
df_2.y.value_counts().plot(kind="bar",
                            color=["green", "orange"])
plt.show()
```

```python
df_2.y.value_counts().plot(kind="bar",
                            color=["blue", "green", "red", "orange"])
plt.show()
```
